Advantages:
Solves vanishing gradient problem of Sigmoid.

Challenges:
Dying ReLU problem: Neurons can become inactive when they always output zero.

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y)
plt.title("ReLU Activation Function")
plt.show()
```

Question 5: Discuss the significance of activation functions in the hidden layers of a neural network.

```python
#Answer
```

They introduce non-linearity, allowing networks to learn complex representations.
Help in gradient propagation during backpropagation.
Ensure efficient training convergence.
Reduce computation costs (e.g., ReLU is computationally cheap).


```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

model = Sequential([
    Dense(10, activation='relu', input_shape=(5,)),
    Dense(10, activation='tanh'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

Question 7: Experiment with different activation functions (e.g., ReLU, Sigmoid, Tanh) in a simple neural network architecture. Compare their effects on convergence and performance.

```python
#Answer
```

Python Example:
We will train a simple neural network using different activation functions and compare their performances.

```python
import tensorflow as tf

def create_model(activation):
    model = Sequential([
        Dense(10, activation=activation, input_shape=(2,)),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

activations = ['relu', 'sigmoid', 'tanh']
for act in activations:
    model = create_model(act)
    print(f"Model with {act} activation function:")
    model.summary()
```
